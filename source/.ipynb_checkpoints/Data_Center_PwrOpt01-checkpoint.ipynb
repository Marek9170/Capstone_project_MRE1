````python
import pandas as pd
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, ConstantKernel, Matern
from scipy.optimize import minimize
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Set style for better-looking plots
plt.style.use('default')
sns.set_palette("husl")

class BayesianDataCenterOptimizer:
    """
    Bayesian Optimization system for Data Center Power Consumption Optimization.
    Uses Gaussian Process regression with Expected Improvement acquisition function.
    """
    
    def __init__(self):
        # Optimization bounds for different parameters
        self.bounds = [
            (20, 100),    # cpu_utilization_percent
            (20, 100),    # ram_utilization_percent  
            (18, 28),     # ambient_temperature_celsius
            (0, 23),      # hour_of_day
            (0, 6),       # day_of_week
            (0, 1),       # peak_hour_flag
            (0, 4),       # server_model_encoded (5 models)
            (0, 7),       # workload_type_encoded (8 types)
        ]
        
        self.X_obs = None
        self.y_obs = None  # Power consumption (to minimize)
        self.efficiency_obs = None  # Power efficiency (to maximize)
        self.gp_power = None
        self.gp_efficiency = None
        self.best_x = None
        self.best_power = np.inf  # Best (lowest) power consumption
        self.best_efficiency = -np.inf  # Best (highest) efficiency
        self.evaluation_history = []
        
        # Server specifications for realistic power modeling
        self.server_specs = {
            0: {'base_power': 140, 'max_power': 550, 'cpu_cores': 8, 'ram_gb': 64},   # R440
            1: {'base_power': 165, 'max_power': 650, 'cpu_cores': 16, 'ram_gb': 128}, # R650
            2: {'base_power': 180, 'max_power': 750, 'cpu_cores': 24, 'ram_gb': 256}, # R750
            3: {'base_power': 150, 'max_power': 800, 'cpu_cores': 32, 'ram_gb': 512}, # C6525
            4: {'base_power': 200, 'max_power': 1400, 'cpu_cores': 48, 'ram_gb': 1024} # R7525
        }
        
        # Workload power multipliers
        self.workload_multipliers = {
            0: 0.3,   # idle
            1: 0.6,   # web_server
            2: 0.75,  # database
            3: 1.2,   # ml_training
            4: 0.9,   # data_analytics
            5: 0.8,   # virtualization
            6: 0.5,   # storage
            7: 1.1    # compute_intensive
        }
        
    def power_consumption_function(self, x):
        """
        Realistic power consumption model based on server configuration and load
        Returns: (power_consumption_watts, power_efficiency_ops_per_watt)
        """
        cpu_util, ram_util, temp, hour, day, peak_flag, server_model, workload = x
        
        # Ensure integer encoding for categorical variables
        server_model = int(round(server_model))
        workload = int(round(workload))
        peak_flag = int(round(peak_flag))
        
        # Clamp to valid ranges
        server_model = max(0, min(4, server_model))
        workload = max(0, min(7, workload))
        
        # Get server specifications
        specs = self.server_specs[server_model]
        base_power = specs['base_power']
        max_power = specs['max_power']
        cpu_cores = specs['cpu_cores']
        ram_gb = specs['ram_gb']
        
        # Calculate variable power based on utilization
        cpu_power_factor = 0.4
        ram_power_factor = 0.15
        other_factor = 0.45
        
        cpu_power = (max_power - base_power) * cpu_power_factor * (cpu_util / 100)
        ram_power = (max_power - base_power) * ram_power_factor * (ram_util / 100)
        other_power = (max_power - base_power) * other_factor * ((cpu_util + ram_util) / 200)
        
        # Apply workload multiplier
        workload_mult = self.workload_multipliers[workload]
        variable_power = (cpu_power + ram_power + other_power) * workload_mult
        
        # Temperature impact (higher temp = higher power due to cooling)
        temp_factor = 1 + (temp - 22) * 0.003  # 0.3% increase per degree above 22¬∞C
        
        # Time of day impact
        if 8 <= hour <= 18:  # Business hours
            time_factor = 1.1
        elif 18 <= hour <= 22:  # Evening peak
            time_factor = 1.05
        else:  # Night/early morning
            time_factor = 0.95
            
        # Peak hour additional load
        peak_factor = 1.15 if peak_flag else 1.0
        
        # Calculate total power
        total_power = (base_power + variable_power) * temp_factor * time_factor * peak_factor
        
        # Add realistic noise
        noise = np.random.normal(0, total_power * 0.02)  # 2% noise
        final_power = max(base_power * 0.8, total_power + noise)
        
        # Calculate power efficiency (operations per watt)
        utilization_score = (cpu_util + ram_util) / 2
        efficiency = (utilization_score * cpu_cores * 0.1) / final_power
        efficiency = max(0.0001, efficiency + np.random.normal(0, efficiency * 0.05))
        
        return final_power, efficiency
    
    def expected_improvement_power(self, x):
        """Expected improvement for power minimization (lower is better)"""
        x = x.reshape(-1, len(self.bounds))
        mu, sigma = self.gp_power.predict(x, return_std=True)
        sigma = sigma.reshape(-1, 1)
        
        f_best = np.min(self.y_obs)  # Best (lowest) power consumption
        xi = 0.01
        
        with np.errstate(divide='warn'):
            # For minimization, we want improvement below current best
            imp = f_best - mu - xi
            Z = imp / sigma
            ei = imp * norm.cdf(Z) + sigma * norm.pdf(Z)
            ei[sigma == 0.0] = 0.0
        
        return ei.flatten()
    
    def expected_improvement_efficiency(self, x):
        """Expected improvement for efficiency maximization (higher is better)"""
        x = x.reshape(-1, len(self.bounds))
        mu, sigma = self.gp_efficiency.predict(x, return_std=True)
        sigma = sigma.reshape(-1, 1)
        
        f_best = np.max(self.efficiency_obs)  # Best (highest) efficiency
        xi = 0.01
        
        with np.errstate(divide='warn'):
            # For maximization, we want improvement above current best
            imp = mu - f_best - xi
            Z = imp / sigma
            ei = imp * norm.cdf(Z) + sigma * norm.pdf(Z)
            ei[sigma == 0.0] = 0.0
        
        return ei.flatten()
    
    def multi_objective_acquisition(self, x):
        """Combined acquisition function balancing power and efficiency"""
        ei_power = self.expected_improvement_power(x)
        ei_efficiency = self.expected_improvement_efficiency(x)
        
        # Weighted combination: prioritize power reduction with efficiency consideration
        power_weight = 0.7
        efficiency_weight = 0.3
        
        # Normalize and combine
        combined_ei = power_weight * ei_power + efficiency_weight * ei_efficiency
        return combined_ei
    
    def optimize_acquisition(self):
        """Find next point to evaluate using acquisition function"""
        best_x = None
        best_acq = -np.inf
        
        # Multiple random starts for global optimization
        for _ in range(50):
            # Random initialization within bounds
            x0 = []
            for low, high in self.bounds:
                x0.append(np.random.uniform(low, high))
            x0 = np.array(x0)
            
            try:
                result = minimize(
                    lambda x: -self.multi_objective_acquisition(x)[0],
                    x0,
                    bounds=self.bounds,
                    method='L-BFGS-B'
                )
                
                if result.success and -result.fun > best_acq:
                    best_acq = -result.fun
                    best_x = result.x
            except Exception as e:
                continue
                
        return best_x if best_x is not None else x0
    
    def generate_initial_samples(self, n_samples=20):
        """Generate strategic initial samples"""
        X_init = []
        
        for i in range(n_samples):
            if i < n_samples // 2:
                # Strategic sampling: focus on energy-efficient configurations
                x = []
                x.append(np.random.uniform(30, 70))  # Moderate CPU utilization
                x.append(np.random.uniform(30, 70))  # Moderate RAM utilization
                x.append(np.random.uniform(20, 24))  # Optimal temperature range
                x.append(np.random.uniform(0, 23))   # Random hour
                x.append(np.random.uniform(0, 6))    # Random day
                x.append(np.random.choice([0, 1]))   # Peak/off-peak
                x.append(np.random.uniform(0, 4))    # Random server model
                x.append(np.random.uniform(0, 7))    # Random workload
            else:
                # Random exploration
                x = []
                for low, high in self.bounds:
                    x.append(np.random.uniform(low, high))
            
            X_init.append(x)
        
        return np.array(X_init)
    
    def run_bayesian_optimization(self, n_iterations=50, n_initial=20):
        """
        Run complete Bayesian optimization for data center power minimization
        """
        print("üöÄ Starting Bayesian Data Center Power Optimization...")
        print("=" * 60)
        
        # Set random seed for reproducibility
        np.random.seed(42)
        
        # Generate initial samples
        print(f"Generating {n_initial} initial samples...")
        X_init = self.generate_initial_samples(n_initial)
        
        # Evaluate initial samples
        y_power_init = []
        y_efficiency_init = []
        
        for i, x in enumerate(X_init):
            power, efficiency = self.power_consumption_function(x)
            y_power_init.append(power)
            y_efficiency_init.append(efficiency)
            
            self.evaluation_history.append({
                'iteration': 0,
                'sample': i,
                'power': power,
                'efficiency': efficiency,
                'config': x.copy()
            })
        
        self.X_obs = X_init
        self.y_obs = np.array(y_power_init)
        self.efficiency_obs = np.array(y_efficiency_init)
        
        # Find initial best
        best_power_idx = np.argmin(self.y_obs)
        best_efficiency_idx = np.argmax(self.efficiency_obs)
        
        self.best_x = self.X_obs[best_power_idx].copy()
        self.best_power = self.y_obs[best_power_idx]
        self.best_efficiency = self.efficiency_obs[best_efficiency_idx]
        
        print(f"Initial best power: {self.best_power:.1f}W")
        print(f"Initial best efficiency: {self.best_efficiency:.4f} ops/W")
        print(f"Best power config: CPU={self.best_x[0]:.1f}%, RAM={self.best_x[1]:.1f}%, Temp={self.best_x[2]:.1f}¬∞C")
        
        # Initialize Gaussian Process models
        power_kernel = ConstantKernel(1.0) * Matern(length_scale=[1.0]*len(self.bounds), nu=2.5) + WhiteKernel(1e-6)
        efficiency_kernel = ConstantKernel(1.0) * Matern(length_scale=[1.0]*len(self.bounds), nu=2.5) + WhiteKernel(1e-6)
        
        self.gp_power = GaussianProcessRegressor(
            kernel=power_kernel,
            alpha=1e-6,
            normalize_y=True,
            n_restarts_optimizer=10,
            random_state=42
        )
        
        self.gp_efficiency = GaussianProcessRegressor(
            kernel=efficiency_kernel,
            alpha=1e-6,
            normalize_y=True,
            n_restarts_optimizer=10,
            random_state=42
        )
        
        # Fit initial models
        self.gp_power.fit(self.X_obs, self.y_obs)
        self.gp_efficiency.fit(self.X_obs, self.efficiency_obs)
        
        print(f"\nüîÑ Starting optimization iterations...")
        
        # Main Bayesian optimization loop
        for iteration in range(n_iterations):
            # Get next point to evaluate
            x_next = self.optimize_acquisition()
            
            if x_next is None:
                print(f"Iter {iteration+1:2d}: Acquisition optimization failed, using random point")
                x_next = []
                for low, high in self.bounds:
                    x_next.append(np.random.uniform(low, high))
                x_next = np.array(x_next)
            
            # Evaluate the new point
            power_next, efficiency_next = self.power_consumption_function(x_next)
            
            # Update observations
            self.X_obs = np.vstack([self.X_obs, x_next])
            self.y_obs = np.append(self.y_obs, power_next)
            self.efficiency_obs = np.append(self.efficiency_obs, efficiency_next)
            
            # Track evaluation history
            self.evaluation_history.append({
                'iteration': iteration + 1,
                'sample': 0,
                'power': power_next,
                'efficiency': efficiency_next,
                'config': x_next.copy()
            })
            
            # Update best solutions
            power_improved = False
            efficiency_improved = False
            
            if power_next < self.best_power:
                old_power = self.best_power
                self.best_power = power_next
                self.best_x = x_next.copy()
                power_improved = True
                improvement = old_power - power_next
                print(f"Iter {iteration+1:2d}: NEW BEST POWER! {power_next:.1f}W (-{improvement:.1f}W) | Eff: {efficiency_next:.4f}")
            
            if efficiency_next > self.best_efficiency:
                self.best_efficiency = efficiency_next
                efficiency_improved = True
                if not power_improved:
                    print(f"Iter {iteration+1:2d}: NEW BEST EFFICIENCY! {efficiency_next:.4f} ops/W | Power: {power_next:.1f}W")
            
            if not power_improved and not efficiency_improved:
                print(f"Iter {iteration+1:2d}: Power: {power_next:.1f}W, Eff: {efficiency_next:.4f} | Best: {self.best_power:.1f}W")
            
            # Refit Gaussian Process models
            self.gp_power.fit(self.X_obs, self.y_obs)
            self.gp_efficiency.fit(self.X_obs, self.efficiency_obs)
            
            # Adaptive bounds refinement in later iterations
            if iteration == n_iterations // 2:
                print(f"\n--- üéØ ADAPTIVE REFINEMENT PHASE ---")
                self._refine_search_bounds()
        
        # Final local optimization
        print(f"\n--- üîç FINAL PRECISION OPTIMIZATION ---")
        self._final_local_search()
        
        return self.get_optimization_results()
    
    def _refine_search_bounds(self):
        """Refine search bounds around promising regions"""
        # Find top 20% performing configurations
        top_indices = np.argsort(self.y_obs)[:len(self.y_obs)//5]
        top_configs = self.X_obs[top_indices]
        
        # Calculate refined bounds based on top performers
        new_bounds = []
        for dim in range(len(self.bounds)):
            values = top_configs[:, dim]
            margin = (values.max() - values.min()) * 0.3  # 30% margin
            center = values.mean()
            
            new_low = max(self.bounds[dim][0], center - margin)
            new_high = min(self.bounds[dim][1], center + margin)
            
            new_bounds.append((new_low, new_high))
        
        self.bounds = new_bounds
        print(f"Refined search bounds around top {len(top_indices)} configurations")
    
    def _final_local_search(self):
        """Final high-precision local search around best solution"""
        best_config = self.best_x.copy()
        best_power = self.best_power
        
        # Create local bounds around best solution
        local_bounds = []
        for i, coord in enumerate(best_config):
            global_low, global_high = self.bounds[i]
            range_size = global_high - global_low
            margin = range_size * 0.1  # 10% of current range
            
            local_low = max(global_low, coord - margin)
            local_high = min(global_high, coord + margin)
            local_bounds.append((local_low, local_high))
        
        # Evaluate additional points in local region
        n_local_samples = 25
        improvements = 0
        
        for _ in range(n_local_samples):
            x_local = []
            for low, high in local_bounds:
                x_local.append(np.random.uniform(low, high))
            x_local = np.array(x_local)
            
            power_local, efficiency_local = self.power_consumption_function(x_local)
            
            # Update observations for completeness
            self.X_obs = np.vstack([self.X_obs, x_local])
            self.y_obs = np.append(self.y_obs, power_local)
            self.efficiency_obs = np.append(self.efficiency_obs, efficiency_local)
            
            # Track in history
            self.evaluation_history.append({
                'iteration': 999,  # Mark as final search
                'sample': improvements,
                'power': power_local,
                'efficiency': efficiency_local,
                'config': x_local.copy()
            })
            
            if power_local < best_power:
                best_power = power_local
                best_config = x_local.copy()
                improvements += 1
                print(f"Local search improvement: {power_local:.1f}W")
        
        if improvements > 0:
            self.best_power = best_power
            self.best_x = best_config
            print(f"Final local search found {improvements} improvements")
        else:
            print("Final local search: no further improvements found")
    
    def get_optimization_results(self):
        """Get comprehensive optimization results"""
        print(f"\n" + "=" * 60)
        print("üèÜ BAYESIAN OPTIMIZATION RESULTS")
        print("=" * 60)
        
        # Best configuration details
        cpu_util, ram_util, temp, hour, day, peak_flag, server_model, workload = self.best_x
        
        server_names = ['R440', 'R650', 'R750', 'C6525', 'R7525']
        workload_names = ['idle', 'web_server', 'database', 'ml_training', 
                         'data_analytics', 'virtualization', 'storage', 'compute_intensive']
        
        print(f"\nüéØ OPTIMAL CONFIGURATION:")
        print(f"Power Consumption: {self.best_power:.1f} watts")
        print(f"Power Efficiency: {self.best_efficiency:.4f} ops/watt")
        print(f"")
        print(f"Configuration Details:")
        print(f"‚îú‚îÄ‚îÄ CPU Utilization: {cpu_util:.1f}%")
        print(f"‚îú‚îÄ‚îÄ RAM Utilization: {ram_util:.1f}%")
        print(f"‚îú‚îÄ‚îÄ Temperature: {temp:.1f}¬∞C")
        print(f"‚îú‚îÄ‚îÄ Hour of Day: {hour:.1f}")
        print(f"‚îú‚îÄ‚îÄ Day of Week: {day:.1f}")
        print(f"‚îú‚îÄ‚îÄ Peak Hour: {'Yes' if peak_flag > 0.5 else 'No'}")
        print(f"‚îú‚îÄ‚îÄ Server Model: {server_names[int(round(server_model))]}")
        print(f"‚îî‚îÄ‚îÄ Workload Type: {workload_names[int(round(workload))]}")
        
        # Calculate potential savings
        baseline_power = np.mean(self.y_obs[:10])  # Average of first 10 samples
        power_reduction = baseline_power - self.best_power
        percentage_reduction = (power_reduction / baseline_power) * 100
        
        print(f"\nüí∞ OPTIMIZATION IMPACT:")
        print(f"‚îú‚îÄ‚îÄ Baseline Power: {baseline_power:.1f}W")
        print(f"‚îú‚îÄ‚îÄ Optimized Power: {self.best_power:.1f}W")
        print(f"‚îú‚îÄ‚îÄ Power Reduction: {power_reduction:.1f}W ({percentage_reduction:.1f}%)")
        print(f"‚îî‚îÄ‚îÄ Annual Cost Savings: ¬£{power_reduction * 8760 * 0.22 / 1000:.0f}")
        
        print(f"\nüìä OPTIMIZATION STATISTICS:")
        print(f"‚îú‚îÄ‚îÄ Total Evaluations: {len(self.evaluation_history)}")
        print(f"‚îú‚îÄ‚îÄ Successful Iterations: {len([h for h in self.evaluation_history if h['power'] < baseline_power])}")
        print(f"‚îî‚îÄ‚îÄ Convergence: {'Excellent' if percentage_reduction > 15 else 'Good' if percentage_reduction > 5 else 'Moderate'}")
        
        return {
            'best_config': self.best_x,
            'best_power': self.best_power,
            'best_efficiency': self.best_efficiency,
            'power_reduction': power_reduction,
            'percentage_reduction': percentage_reduction,
            'annual_savings_gbp': power_reduction * 8760 * 0.22 / 1000,
            'total_evaluations': len(self.evaluation_history),
            'evaluation_history': self.evaluation_history
        }
    
    def plot_optimization_progress(self):
        """Visualize the optimization progress"""
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
        fig.suptitle('Bayesian Optimization Progress for Data Center Power Optimization', 
                    fontsize=16, fontweight='bold')
        
        # Extract data for plotting
        iterations = [h['iteration'] for h in self.evaluation_history]
        powers = [h['power'] for h in self.evaluation_history]
        efficiencies = [h['efficiency'] for h in self.evaluation_history]
        
        # 1. Power consumption over iterations
        ax1.plot(iterations, powers, 'b-', alpha=0.6, label='Evaluated Points')
        
        # Running minimum
        running_min = []
        current_min = float('inf')
        for power in powers:
            if power < current_min:
                current_min = power
            running_min.append(current_min)
        
        ax1.plot(iterations, running_min, 'r-', linewidth=3, label='Best Power')
        ax1.set_title('Power Consumption Optimization', fontweight='bold')
        ax1.set_xlabel('Iteration')
        ax1.set_ylabel('Power Consumption (W)')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # 2. Power efficiency over iterations
        ax2.plot(iterations, efficiencies, 'g-', alpha=0.6, label='Evaluated Points')
        
        # Running maximum for efficiency
        running_max = []
        current_max = -float('inf')
        for eff in efficiencies:
            if eff > current_max:
                current_max = eff
            running_max.append(current_max)
        
        ax2.plot(iterations, running_max, 'purple', linewidth=3, label='Best Efficiency')
        ax2.set_title('Power Efficiency Optimization', fontweight='bold')
        ax2.set_xlabel('Iteration')
        ax2.set_ylabel('Power Efficiency (ops/watt)')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        # 3. Power vs Efficiency scatter plot
        colors = np.array(iterations)
        scatter = ax3.scatter(powers, efficiencies, c=colors, cmap='viridis', alpha=0.6)
        
        # Find best efficiency corresponding to best power (or close to it)
        best_power_entries = [h for h in self.evaluation_history if abs(h['power'] - self.best_power) < 1.0]
        if best_power_entries:
            best_entry = best_power_entries[0]
            best_eff_for_plot = best_entry['efficiency']
        else:
            best_eff_for_plot = self.best_efficiency
            
        ax3.scatter([self.best_power], [best_eff_for_plot], c='red', s=200, 
                   marker='*', label='Best Solution', edgecolors='black', linewidth=2)
        ax3.set_title('Power vs Efficiency Trade-off', fontweight='bold')
        ax3.set_xlabel('Power Consumption (W)')
        ax3.set_ylabel('Power Efficiency (ops/watt)')
        ax3.legend()
        ax3.grid(True, alpha=0.3)
        plt.colorbar(scatter, ax=ax3, label='Iteration')
        
        # 4. Configuration parameter evolution
        configs = np.array([h['config'] for h in self.evaluation_history])
        param_names = ['CPU%', 'RAM%', 'Temp¬∞C', 'Hour', 'Day', 'Peak', 'Server', 'Workload']
        
        # Show evolution of key parameters
        ax4.plot(iterations, configs[:, 0], label='CPU Util %', alpha=0.7)
        ax4.plot(iterations, configs[:, 1], label='RAM Util %', alpha=0.7)
        ax4.plot(iterations, configs[:, 2] * 4, label='Temperature (√ó4)', alpha=0.7)  # Scale for visibility
        
        # Mark best configuration
        best_iters = [h['iteration'] for h in self.evaluation_history if abs(h['power'] - self.best_power) < 0.1]
        if best_iters:
            best_iter = best_iters[0]
            ax4.axvline(x=best_iter, color='red', linestyle='--', alpha=0.7, label='Best Config')
        else:
            # Fallback: find iteration with minimum power
            min_power_idx = np.argmin([h['power'] for h in self.evaluation_history])
            best_iter = self.evaluation_history[min_power_idx]['iteration']
            ax4.axvline(x=best_iter, color='red', linestyle='--', alpha=0.7, label='Best Config')
        
        ax4.set_title('Parameter Evolution', fontweight='bold')
        ax4.set_xlabel('Iteration')
        ax4.set_ylabel('Parameter Value')
        ax4.legend()
        ax4.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig('bayesian_optimization_progress.png', dpi=300, bbox_inches='tight')
        print("üìà Optimization progress plots saved as: bayesian_optimization_progress.png")
        plt.show()
        
        return fig

# Main execution function
def run_bayesian_datacenter_optimization():
    """Main function to run Bayesian optimization for data center power efficiency"""
    
    optimizer = BayesianDataCenterOptimizer()
    
    # Run the optimization
    results = optimizer.run_bayesian_optimization(n_iterations=50, n_initial=25)
    
    # Generate visualizations
    optimizer.plot_optimization_progress()
    
    print(f"\n‚úÖ BAYESIAN OPTIMIZATION COMPLETE!")
    print(f"üìä Check the visualization plots for detailed analysis")
    print(f"üîß Best configuration saved and ready for implementation")
    
    return optimizer, results

if __name__ == "__main__":
    optimizer, results = run_bayesian_datacenter_optimization()
````
